# Nested Pie Chart

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

## Example 1

```python

import matplotlib.pyplot as plt
import numpy as np

size = 0.3

# Create a figure and axes
fig, ax = plt.subplots()

# Define the values for the pie chart
vals = np.array([[40., 32.], [47., 40.], [79., 20.]])
# The values represent the data for each pie slice

# Get the color map
cmap = plt.get_cmap("tab20c")

# Define the colors for the outer pie slices
outer_colors = cmap(np.arange(3) * 4)

# Define the colors for the inner pie slices
inner_colors = cmap([1, 2, 5, 6, 9, 10])

# Plot the outer pie chart
ax.pie(vals.sum(axis=1), radius=1, colors=outer_colors, wedgeprops=dict(width=0.3, edgecolor='0'))
# This creates the outer pie chart using the sum of values along each row
# The radius is set to 1, colors are set to outer_colors, and the wedge properties are customized

# Plot the inner pie chart
ax.pie(vals.flatten(), radius=1-size, colors=inner_colors, wedgeprops=dict(width=0.7, edgecolor='w'))
# This creates the inner pie chart using the flattened values from the array
# The radius is set to 1-size (where size is a variable), colors are set to inner_colors, and the wedge properties are customized

# Set the aspect ratio and title of the plot
ax.set(aspect="equal", title='Pie plot with `ax.pie`')

# Display the plot
plt.show()
# This displays the pie chart plotm
```

## Example 2

```python
import matplotlib.pyplot as plt

names='A', 'B', 'C', 'D',
size=[12,11,3,30]

fig = plt.figure()
fig.patch.set_facecolor('grey')

plt.rcParams['text.color'] = 'white'

my_circle=plt.Circle( (0,0), 0.7, color='grey')

plt.title('Donut Chart')
plt.pie(size, labels=names)
p = plt.gcf()
p.gca().add_artist(my_circle)
plt.show()
```
